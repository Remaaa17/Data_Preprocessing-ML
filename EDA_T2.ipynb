```python
# Replace '?' with NaN
data.replace('?', pd.NA, inplace=True)

# Calculate mode for each column
workclass_mode = data['workclass'].mode()[0]
occupation_mode = data['occupation'].mode()[0]
native_country_mode = data['native-country'].mode()[0]

# Fill missing values with mode
data['workclass'].fillna(workclass_mode, inplace=True)
data['occupation'].fillna(occupation_mode, inplace=True)
data['native-country'].fillna(native_country_mode, inplace=True)

print(data)
```

>* **replaced {?} with null value and fill these  values withe the most frequent word in each col**

```python
def convert_marital_status(status):
    if status in ['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse']:
        return 'married'
    elif status in ['Never-married', 'Separated', 'Widowed']:
        return 'single'
    else:
        return 'divorced'

data['marital-status'] = data['marital-status'].apply(convert_marital_status)
```

>* **change marital_status To (married , single, divorced) and remove unnessary other states**

```python
from operator import imod
import seaborn as sns
import matplotlib.pyplot as plt
import math

def plot_bivariate_bar(dataset, hue, cols=5, width=20, height=15, hspace=0.2, wspace=0.5):
    dataset = dataset.select_dtypes(include=['object'])
    sns.set_style('whitegrid')  # Use seaborn's set_style instead of plt.style.use
    fig = plt.figure(figsize=(width, height))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=wspace, hspace=hspace)
    rows = math.ceil(float(dataset.shape[1]) / cols)
    for i, column in enumerate(dataset.columns):
        ax = fig.add_subplot(rows, cols, i + 1)
        ax.set_title(column)
        if dataset.dtypes[column] == object:
            g = sns.countplot(y=column, hue=hue, data=dataset, palette='BuPu')  # Assign palette directly
            # Extract the text from the Text objects before slicing
            ax.set_yticklabels([label.get_text()[:10] for label in ax.get_yticklabels()])  # Truncate labels directly

    plt.show()

# Assuming 'data' is your DataFrame with relevant columns
bivariate_df = data[['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'income']]
plot_bivariate_bar(bivariate_df, hue='income', cols=2, width=20, height=15, hspace=0.4, wspace=0.5)
```

```python
plt.figure(figsize=(5, 2))
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='BuPu', fmt=".2f", linewidths=.5)
plt.show()
```

```python
# Identify positive and negative correlations
positive_corr = []
negative_corr = []

# Loop through correlation matrix
for col1 in correlation_matrix.columns:
    for col2 in correlation_matrix.columns:
        if col1 != col2:
            corr_value = correlation_matrix.loc[col1, col2]
            if corr_value > 0:
                positive_corr.append((col1, col2, corr_value))
            elif corr_value < 0:
                negative_corr.append((col1, col2, corr_value))

# Print positive correlations
print("Positive Correlations:")
for col1, col2, corr_value in positive_corr:
    print(f"{col1} <-> {col2}: {corr_value}")

# Print negative correlations
print("\nNegative Correlations:")
for col1, col2, corr_value in negative_corr:
    print(f"{col1} <-> {col2}: {corr_value}")
```

###Feature Scaling & Handling Outliers

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=data['age'], color = '#6A0DAD')
plt.title('Age Feature Before Removing Outliers')
plt.show()
```

>* **Age have outliers**

```python
data.info()
```

```python
# Calculate quartiles and IQR
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['age'] < lower_bound) | (data['age'] > upper_bound)]

# Drop outliers from the original DataFrame
data.drop(outliers.index, inplace=True)
```

```python
data.info()
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=data['age'],color = '#6A0DAD')
plt.title('Age Feature After Removing the Outliers')
plt.show()
```
